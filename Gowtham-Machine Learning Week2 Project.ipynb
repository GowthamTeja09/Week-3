Step Two

```python
import pdfplumber
import pandas as pd

# Open and read the PDF file
with pdfplumber.open(pdf_path) as pdf:
    table_data = []

    # Loop through all pages to extract tables
    for page in pdf.pages:
        tables = page.extract_table()
        if tables:
            table_data.extend(tables)

# Convert the extracted data into a DataFrame
df = pd.DataFrame(table_data)

# Display the first few rows
df.head()
# Remove any empty rows or columns
df.dropna(how='all', inplace=True)

# Reset column names if needed (modify based on extracted data)
df.columns = df.iloc[0]  # Set first row as header
df = df[1:].reset_index(drop=True)

# Save the cleaned data to a CSV file
csv_path = "/mnt/data/toyota_corolla_data.csv"
df.to_csv(csv_path, index=False)
print(f"Data saved to: {csv_path}")
```

Step Three

```python
df = pd.read_csv(csv_path)

df.info()

df.head()

df.describe()

df.isnull().sum()

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.histplot(df["Price"], bins=30, kde=True)
plt.title("Price Distribution of Toyota Corolla Cars")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Age_08_04"], y=df["Price"])
plt.title("Age vs. Price")
plt.xlabel("Age (in months)")
plt.ylabel("Price")
plt.show()
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Fuel_Type"], y=df["Price"])
plt.title("Fuel Type vs. Price")
plt.xlabel("Fuel Type")
plt.ylabel("Price")
plt.show()
```
